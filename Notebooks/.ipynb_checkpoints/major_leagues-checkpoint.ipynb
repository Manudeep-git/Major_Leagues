```python
#creating a new column team_won
df=df.assign(Team_won=[i for i in range(df.shape[0])])
```

```python
#Determining which Team won
def team_won(cols):
    score1=cols[0]
    score2=cols[1]
    team1=cols[2]
    team2=cols[3]
    
    if score1>score2:
        return team1
    elif score1<score2:
        return team2
    else:
        return "Tie"
```

```python
def winning_goals(cols):
    score1=cols[0]
    score2=cols[1]
    
    if score1>score2:
        return score1
    elif score1<score2:
        return score2
```

```python
df['Team_won']=df[['score1','score2','team1','team2','Team_won']].apply(team_won,axis=1)
df['Winning_Goals']= df[['score1','score2']].apply(winning_goals,axis=1)
```

```python
df.head()
```

```python
sns.countplot(x="Winning_Goals",data=df)
```

Majority of teams have won matches with 2 goals, followed by 3 goals and 1 goal, very rarely there were teams who scored more than 4 golas

### Teams with total goals

```python
x = df.groupby("team1")["score1"].sum().reset_index()
x = x.rename(columns={'team1':"team", 'score1':"goals"})
y = data.groupby("team2")["score2"].sum().reset_index()
y = y.rename(columns={'team2':"team", 'score2':"goals"})
goals = pd.concat([x,y])
goals = goals.groupby("team")["goals"].sum().reset_index().sort_values(by = "goals",ascending = False)
plt.figure(figsize=(9,14))
ax = sns.barplot(x="goals",y="team",
                 data=goals[:20],palette="rainbow",
                linewidth = 1,edgecolor = "k"*20)
for i,j in enumerate(goals["goals"][:20]):
    ax.text(.3,i,j,weight="bold",color = "k",fontsize =12)
plt.title("Teams with highest total goals ")
plt.show()
```

### 3 Feature Engineering

#### 3.1 Variance Threshold

```python
# Using Variance threshold to select variant features and drop constant features
X=df.drop(columns=['score1','score2','date','league_id','team1','team2','league','Team_won'],axis=1)
Y=df[['score1','score2']]
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(X)
```

```python
var_thres.get_support()
```

```python
##Finding non constant features
sum(var_thres.get_support())
```

Variance threshold doesn't give much information as there are no columns with zero variance

#### 3.3 Using Feature Transformation

```python
#changing "league", "team1", "team2" to category
df1["league"] = df1["league"].astype('category')
df1["team1"] = df1["team1"].astype('category')
df1["team2"] = df1["team2"].astype('category')
df1['Team_won'] = df1['Team_won'].astype('category')
```

```python
df_league=pd.get_dummies(df1['league'])
df_team1=pd.get_dummies(df1, prefix=['Team1_'],columns=['team1'],drop_first=True)
df_team2=pd.get_dummies(df1, prefix=['Team2_'],columns=['team2'],drop_first=True)
df_team_won= pd.get_dummies(df1, columns=['Team_won'],drop_first=True)
df1=pd.concat([df1,df_league,df_team1,df_team2],axis=1)
```

```python
df1.drop(columns=['team1','team2','Team_won','league','Winning_Goals'],inplace=True)
```
